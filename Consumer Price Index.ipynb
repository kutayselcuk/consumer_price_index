```python
#function for getting rows which I want(rows which are common between two data set)
def merge_the_dataframes(leftFrame, rightFrame, commonProducts):
    
    preResult = pd.DataFrame()
    result = pd.DataFrame()
    
    tüfeProductNameTrSeries = leftFrame['tüfe_Name_Tr']
    tüfeProductCodeSeries = leftFrame['tüfe_Code']
    tüfeProductRatioSeries = leftFrame['tüfe_Ratio']
    tüfeProductNameEngSeries = leftFrame['tüfe_Name_Eng']
    tüfeRelatedWordsSeries = leftFrame["tüfe_Related_Words"]
    tüfeRelatedWeightsSeries = leftFrame["tüfe_Related_Weights"]
    tüfeExcludedWordsSeries = leftFrame["tüfe_Excluded_Words"]
    
    pickleItemNamesSeries = rightFrame["pickle_Name"]
    
    for x in leftFrame.index:
        currentItemTüfeNameTrForContains = " " + tüfeProductNameTrSeries[x] + " "
        #also can be used as tüfeProductNameTrSeries[x], exp. to take products which are include "pirinçli", but this time code take all sütaş products when we take "süt"

        preResult = rightFrame[rightFrame['pickle_Name'].astype(str).str.contains(currentItemTüfeNameTrForContains)]
        
        
        preResult.insert(3, "tüfe_Name_Tr", tüfeProductNameTrSeries[x], True)
        preResult.insert(4, "tüfe_Related_Words", tüfeRelatedWordsSeries[x], True)
        preResult.insert(5, "tüfe_Ratio", tüfeProductRatioSeries[x], True)
        preResult.insert(6, "tüfe_Name_Eng", tüfeProductNameEngSeries[x], True)
        preResult.insert(7, "tüfe_Code", tüfeProductCodeSeries[x], True)
        preResult.insert(8, "tüfe_Related_Weights", tüfeRelatedWeightsSeries[x], True)
        preResult.insert(9, "tüfe_Excluded_Words", tüfeExcludedWordsSeries[x], True)
        
        
        result = result.append(preResult)
    
    return result
```

```python
#common product DataFrame and arrange it for process
commonProducts = pd.DataFrame()
commonProducts = merge_the_dataframes(tüfe, pickle, commonProducts)
commonProducts = commonProducts.set_index(np.arange(len(commonProducts["pickle_Name"])))
commonProducts
```

```python
def getWeightInfo(nameColumn):
    
    weightsAndUnits = ["" for i in range(len(nameColumn))]
    weightsOfItems = nameColumn.str.split().str[-2]
    unitsOfItems = nameColumn.str.split().str[-1]
    
    for x in range(len(nameColumn)):
        if(weightsOfItems[x].isalpha()):
            weightsAndUnits[x] += "1 "
        else:
            weightsAndUnits[x] += weightsOfItems[x] + " "
            
            #cutting the weight from name
            text = nameColumn[x]
            text = text.rsplit(' ', 2)[0] + " " +text.rsplit(' ', 2)[2]
            nameColumn[x] = text
            
        weightsAndUnits[x] += unitsOfItems[x]
        
        #cutting the unit from name
        text2 = nameColumn[x]
        text2 = text2.rsplit(' ', 1)[0]
        nameColumn[x] = text2
        
        
        
    return weightsAndUnits
```

```python
def categorizeForWeightsOfItems(df):
    
    result = pd.DataFrame(columns=df.columns)
    
    tüfeRelatedWeightsSeries = df["tüfe_Related_Weights"]
    weightsOfItemsSeries = df["weights_Of_Items"]
    
    for x in df.index:
        
        currentItemWeightWithUnit = weightsOfItemsSeries[x]
        currentItemWeight = currentItemWeightWithUnit.split(" ")[0]

        currentListOfRelatedWeights = tüfeRelatedWeightsSeries[x]
        splitedList = currentListOfRelatedWeights.split(",")
        #print(splitedList)
        
        
        if currentItemWeight in splitedList:
            
            df["tüfe_Related_Weights"][x] = currentItemWeight
            
            result = result.append(df.iloc[x])
        elif "*" in splitedList:
            
            result = result.append(df.iloc[x])
            
    result = result.set_index(np.arange(len(result["pickle_Name"])))
                
    return result  
```
